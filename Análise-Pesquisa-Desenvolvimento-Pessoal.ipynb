```python
# Análise da Pesquisa sobre Desenvolvimento Pessoal

import pandas as pd
import pandas_profiling
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%config InlineBackend.figure_format = 'svg'
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
df = pd.read_csv('/Users/rauferibeiro/Desenvolvimento.csv')
df.head()
```

```python
# Eliminando as colunas que não serão utilizadas na análise

df.drop('Unnamed: 15', inplace=True, axis=1)
df.drop('Unnamed: 16', inplace=True, axis=1)
df.drop('Carimbo de data/hora', inplace=True, axis=1)
df.head()
```

```python
#Estatísticas básicas a respeito das variáveis

# df.profile_report() -- Mostra estatísticas completas, gráficos e outras infos interessantes

df.describe()
```

```python
# Calcula o número de dados faltantes em cada coluna

def num_missing(x):
  return sum(x.isnull())

print("Missing values per column:")
print(df.apply(num_missing, axis=0)) #axis=0 define que a função deve ser aplicada em cada coluna
```

```python
#Elimina as colunas com valores faltantes no dataset

df = df.dropna()
```

```python
# Formatando a coluna Idade para retirar lixos vindos da pesquisa

df['Idade'] = df['Idade'].str.extract('(\d{2})', expand=False)
df['Idade'] = df['Idade'].astype('int64')
```

```python
df.head()
```

```python
# Definindo o tamanho do gráfico
plt.rcParams['figure.figsize'] = (8,5)

# Criando a lista de cores
colors = ['#ffb3e6','#66b3ff']

# Gráfico de Pizza da distribuição de Gênero
fatias = plt.pie(df.Genero.value_counts(), colors = colors, explode = (0.1,0), autopct='%1.1f%%', startangle=90)
plt.legend(fatias, labels=['Femino','Masculino'], bbox_to_anchor = (1.05,1))
plt.title('Gênero')
plt.show()
```

```python
# Distribuição da Idade dos respondentes 

plt.hist(df.Idade)
```

```python
df['Faixas_Idade'] = pd.cut(df.Idade,\
                   bins=[0, 18, 29, 39, 49, 59, 999],\
                   labels=['0-19', '20-29', '30-39', '40-49', '50-59', '60+'],\
                   include_lowest=True)
df.head()
```
